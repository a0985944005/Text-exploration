```python
from gensim import corpora,models,similarities
from gensim import matutils

###step1 將經過段詞結果的三個文本 DOC1、2、3作為於料來源

documents = [['我','聽見','雨滴','落在','青青','草地'],['我','聽見','遠方','下課','的','鐘聲','響起'],['我','沒有','聽見','你','的','聲音']]

###step2 將list中的字詞轉換為word to id 映射字典bag of words 詞袋

dictionary = corpora.Dictionary(documents)
print(dictionary)
print(dictionary.token2id)

###step3 江文本統計轉化成依照字典中各詞向的編號語詞類
#tuple style:(token_id,token_count)
corpus=[dictionary.doc2bow(text)for text in documents]
print(corpus)

#(0,1)表示我這個詞出現在DOC1一次

###step4 將語料庫轉乘字詞-文本舉證 (TDM)
#將複雜的corpus轉換成TDM稀疏矩陣

print(len(dictionary)) #詞向數量

numpy_matrix=matutils.corpus2dense(corpus,num_terms=len(dictionary)) #num_terms=tokens數量
print(numpy_matrix)
```

```python
import jieba
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer

##step1:讀取語料庫，進行斷詞
corpus=[]
with open('input.txt','r',encoding='utf8') as f:
    for line in f:
        corpus.append("|".join(jieba.cut(line.strip())))
print(corpus)


#step2:運算取得tf-idf matrix
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
#其中列為檔、行為字詞，表示對應的tf-idf value
tfidf_matrix=tfidf.toarray()
print(tfidf.shape)   #(幾個分檔，幾個詞項)

#取的所有檔字詞特徵
wordfeatures = vectorizer.get_feature_names()
print(wordfeatures)

#列印每份文件中，tf-idf value 大於0的關鍵字及對應值
for i in range(len(corpus)):
    print('-----Document%d-----'%(i))
    for j in range(len(wordfeatures)):
        if tfidf[i,j]>0:
            print(wordfeatures[j],tfidf[i,j])
```
